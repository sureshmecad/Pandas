<h2 style="color:blue" align="left"> 1. Imputer </h2>

```python
from sklearn.impute import SimpleImputer
```

```python
df_imputer = SimpleImputer(missing_values=np.nan, strategy='mean', verbose=0)
```

```python
df_imputer = df_imputer.fit(df[1:,1:8])
df[1:,1:8] = df_imputer.transform(df[1:,1:8])
```

```python
df
```

```python
y = pd.DataFrame(df)
y
```

- The default value for this method is the mean value of imputation.

- That is, each NaN value of column 0 is filled with the mean of the non-NaN data points corresponding to column 0.
