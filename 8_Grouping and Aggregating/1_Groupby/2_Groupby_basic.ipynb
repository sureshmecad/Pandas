## Groupby

## 1. Dataset : nba

```python
import pandas as pd
```

```python
df = pd.read_csv("nba.csv")
df.head()
```

<h3 style="color:blue" align="left"> 1. Use groupby( ) function to group the data based on the “Team” </h3>

```python
df['Team'].value_counts().sort_index(ascending=True)
```

```python
df['Team'].nunique()
```

```python
df['Team'].shape
```

```python
gk = df.groupby('Team')

# Print the first value in each group
gk.first()
```

### Finding the values contained in the "Boston Celtics" group

```python
gk.get_group('Boston Celtics')
```

<h3 style="color:blue" align="left"> 2. Use groupby( ) function to form groups based on more than one category (i.e. Use more than one column to perform the splitting) </h3>

```python
df['Position'].value_counts().sort_index(ascending=True)
```

```python
df['Position'].nunique()
```

```python
df['Position'].shape
```

```python
# First grouping based on "Team" 
# Within each team we are grouping based on "Position"

gkk = df.groupby(['Team', 'Position']) 
  
# Print the first value in each group 
gkk.first() 
```

## 2. Dataset : Titanic

```python
train = pd.read_csv("titanic.csv")
train.head()
```

```python
train.groupby('Survived').mean()
```

```python
train.groupby('Sex').mean()
```

```python
train.groupby(['Sex','Survived']).count()
```

```python
train.groupby(['Sex','Survived'])['Age'].count()
```

```python
train.groupby(['Pclass','Survived']).count()
```
