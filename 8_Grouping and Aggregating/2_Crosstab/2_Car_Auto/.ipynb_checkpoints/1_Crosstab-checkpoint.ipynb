```python
import pandas as pd
import seaborn as sns
```

```python
train = pd.read_csv("train.csv")
train.head()
```

## 2. Pclass:- Ordinal Feature

```python
train['Pclass'].value_counts()
```

```python
pd.crosstab(train['Pclass'], train['Survived'], margins=True).style.background_gradient(cmap='summer_r')
```

```python
orange = sns.light_palette("orange", as_cmap=True)
pd.crosstab(train['Pclass'], train['Survived'], margins=True).style.background_gradient(cmap=orange)
```

------------------

```python
pd.crosstab([train['Sex'], train['Survived']], train['Pclass'], margins=True).style.background_gradient(cmap='summer_r')
```

```python
pd.crosstab([train['Sex'], train['Survived']], train['Pclass'], margins=False).style.background_gradient(cmap='summer_r')
```

```python
pd.crosstab([train['Sex'], train['Survived']], train['Pclass'], margins=True, margins_name="Total").style.background_gradient(cmap='summer_r')
```

---------------

### 4. Embarked:- Categorical Value

```python
pd.crosstab([train['Embarked'], train['Pclass']], [train['Sex'], train['Survived']], margins=True).style.background_gradient(cmap='summer_r')
```

### 5. SibSip:- Discrete Feature

```python
pd.crosstab([train['SibSp']], train['Survived']).style.background_gradient(cmap='summer_r')
```

```python
pd.crosstab(train['SibSp'], train['Pclass']).style.background_gradient(cmap='summer_r')
```

### 6. Parch

```python
pd.crosstab(train['Parch'], train['Pclass']).style.background_gradient(cmap='summer_r')
```

```python
train['Age_band']=0
train.loc[train['Age']<=16,'Age_band']=0
train.loc[(train['Age']>16)&(train['Age']<=32),'Age_band']=1
train.loc[(train['Age']>32)&(train['Age']<=48),'Age_band']=2
train.loc[(train['Age']>48)&(train['Age']<=64),'Age_band']=3
train.loc[train['Age']>64,'Age_band']=4
train.head(2)
```

```python
# checking the number of passenegers in each band
train['Age_band'].value_counts().to_frame().style.background_gradient(cmap='summer')
```
