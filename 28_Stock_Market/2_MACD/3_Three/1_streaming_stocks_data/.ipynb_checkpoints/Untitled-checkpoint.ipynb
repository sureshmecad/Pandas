```python
# Read Data
import numpy as np
import pandas as pd

# Visualization

# Matplotlib and Seaborn
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

# Plotly
import plotly
import plotly.express as px
import plotly.graph_objs as go
import plotly.figure_factory as ff
from plotly.subplots import make_subplots
from plotly.offline import plot, iplot, init_notebook_mode

import os, time
import datetime
import pytz
from datetime import datetime
timezone_str = 'Asia/Kolkata'
localtz = pytz.timezone(timezone_str)


# style
plt.style.use("fivethirtyeight")
sns.set_style("darkgrid")
```

```python
# Loading Stocks data
stocks = pd.read_csv("streaming_stocks_data.csv")
stocks.columns = ['timestamp','last_price','volume','scrip_id']
stocks.head()
```

```python
def get_stocks_close(stocks, scripid):
    stocks_close = stocks.loc[stocks['scrip_id'] == scripid]

    stocks_close.reset_index(inplace=True)

    stocks_close.drop(['index'], axis=1, inplace=True)

    for i in range(len(stocks_close)):
        stocks_close['timestamp'][i] = stocks_close['timestamp'][i][:-6]

    stocks_close['timestamp'] = pd.to_datetime(stocks_close['timestamp'])

    stocks_close.set_index('timestamp',inplace=True)

    return stocks_close
```

```python
streaming = get_stocks_close(stocks, 14)
print(streaming.shape)
streaming.head()
```

```python
# Visually show the stock price
plt.figure(figsize=(15,10))
plt.plot(streaming['last_price'], label='Last Price')
plt.title('Close Price History')
plt.xlabel('Timestamp')
plt.ylabel('Price')
plt.xticks(rotation=45)
plt.show()
```

```python
def get_stocks_closeprice(stocks_close, name_of_co):
    stocks_close_co = pd.DataFrame()
    stocks_close_co[name_of_co] = stocks_close['last_price'].values
    return stocks_close_co

stocks_close_rel = get_stocks_closeprice(streaming, "streaming")
stocks_close_rel.head()
```

```python
# Calculate the three moving averages
# Calculate the short / fast exponential moving average (EMA)
ShortEMA = streaming['last_price'].ewm(span=5, adjust=False).mean()

# Calculate the Middle / Medium exponential moving average (EMA)
MiddleEMA = streaming['last_price'].ewm(span=21, adjust=False).mean()

# Calculate the long/slow exponential moving average (EMA)
LongEMA = streaming['last_price'].ewm(span=60, adjust=False).mean()
```

```python
# Visualize the Last Price and the exponential moving averages
plt.figure(figsize=(15,12))
plt.title('Last Price', fontsize=18)
plt.plot(streaming['last_price'], label='Last Price', color='blue')
plt.plot(ShortEMA, label='Short/Fast EMA', color='red')
plt.plot(MiddleEMA, label='Middle/Medium EMA', color='orange')
plt.plot(LongEMA, label='Long/Slow EMA', color='green')
plt.xticks(rotation=45)
plt.xlabel('Timestamp', fontsize=18)
plt.ylabel('Last Price', fontsize=18)
plt.legend(loc='upper left')
plt.show()
```

```python
def buy_sell_function(data):
    buy_list = []
    sell_list = []
    flag_long = False
    flag_short = False
      
    for i in range(0, len(data)):
      if data['Middle'][i] < data['Long'][i] and data['Short'][i] < data['Middle'][i] and flag_long == False and flag_short == False:
        buy_list.append(data['last_price'][i])
        sell_list.append(np.nan)
        flag_short = True
      elif flag_short == True and data['Short'][i] > data['Middle'][i]:
        sell_list.append(data['last_price'][i])
        buy_list.append(np.nan)
        flag_short = False
      elif data['Middle'][i] > data['Long'][i] and data['Short'][i] > data['Middle'][i] and flag_long == False and flag_short == False:
        buy_list.append(data['last_price'][i])
        sell_list.append(np.nan)
        flag_long = True
      elif flag_long == True and data['Short'][i] < data['Middle'][i]:
        sell_list.append(data['last_price'][i])
        buy_list.append(np.nan)
        flag_long = False
      else:
        buy_list.append(np.nan)
        sell_list.append(np.nan)

    return (buy_list, sell_list)
```

```python
# get signal df i.e df telling when to buy/sel

def get_signal(stocks_close_co, co_name, ShortEMA, MiddleEMA, LongEMA):
    signal = pd.DataFrame(index=stocks_close_co[co_name].index)
    signal[co_name] = stocks_close_co[co_name]
    
    signal['ShortEMA'] = ShortEMA[co_name]
    signal['MiddleEMA'] = MiddleEMA[co_name]
    signal['LongEMA'] = LongEMA[co_name]
    
    x = buy_sell(signal, co_name)
    signal['Buy_Signal_Price'] = x[0]
    signal['Sell_Signal_Price'] = x[1]
    
    return signal

signal = get_signal(stocks_close_rel, "streaming", ShortEMA, MiddleEMA, LongEMA)
signal.head()    
```
