- This program uses the **Three Moving Average Convergence / Divergence (MACD)** crossover to determine when to buy and sell stock.

```python
# import the libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')
```

```python
# Loading Stocks data
df = pd.read_csv("streaming_stocks_data.csv")
df.columns = ['timestamp','last_price','volume','scrip_id']
df.head()
```

```python
df = df.set_index(pd.DatetimeIndex(df['timestamp'].values))
df.head()
```

```python
df1 = df[0:374]
df1.head()
```

```python
# Visually show the stock price
plt.figure(figsize=(15,10))
plt.plot(df1['last_price'], label='Last Price')
plt.title('Close Price History')
plt.xlabel('Timestamp')
plt.ylabel('Price')
plt.xticks(rotation=45)
plt.show()
```

```python
# Calculate the three moving averages
# Calculate the short / fast exponential moving average (EMA)
ShortEMA = df1['last_price'].ewm(span=5, adjust=False).mean()

# Calculate the Middle / Medium exponential moving average (EMA)
MiddleEMA = df1['last_price'].ewm(span=21, adjust=False).mean()

# Calculate the long/slow exponential moving average (EMA)
LongEMA = df1['last_price'].ewm(span=60, adjust=False).mean()
```

```python
# Visualize the Last Price and the exponential moving averages
plt.figure(figsize=(15,12))
plt.title('Last Price', fontsize=18)
plt.plot(df1['last_price'], label='Last Price', color='blue')
plt.plot(ShortEMA, label='Short/Fast EMA', color='red')
plt.plot(MiddleEMA, label='Middle/Medium EMA', color='orange')
plt.plot(LongEMA, label='Long/Slow EMA', color='green')
plt.xticks(rotation=45)
plt.xlabel('Timestamp', fontsize=18)
plt.ylabel('Last Price', fontsize=18)
plt.legend(loc='upper left')
plt.show()
```

```python
# Add the exponential moving averages to the dataset
df1['Short'] = ShortEMA
df1['Middle'] = MiddleEMA
df1['Long'] = LongEMA
df1.head()
```

```python
def buy_sell_function(data):
    buy_list = []
    sell_list = []
    flag_long = False
    flag_short = False
      
    for i in range(0, len(data)):
      if data['Middle'][i] < data['Long'][i] and data['Short'][i] < data['Middle'][i] and flag_long == False and flag_short == False:
        buy_list.append(data['last_price'][i])
        sell_list.append(np.nan)
        flag_short = True
      elif flag_short == True and data['Short'][i] > data['Middle'][i]:
        sell_list.append(data['last_price'][i])
        buy_list.append(np.nan)
        flag_short = False
      elif data['Middle'][i] > data['Long'][i] and data['Short'][i] > data['Middle'][i] and flag_long == False and flag_short == False:
        buy_list.append(data['last_price'][i])
        sell_list.append(np.nan)
        flag_long = True
      elif flag_long == True and data['Short'][i] < data['Middle'][i]:
        sell_list.append(data['last_price'][i])
        buy_list.append(np.nan)
        flag_long = False
      else:
        buy_list.append(np.nan)
        sell_list.append(np.nan)

    return (buy_list, sell_list)
```

```python
# Add the buy and sell signals to the dataset
df1['Buy'] = buy_sell_function(df1)[0]
df1['Sell'] = buy_sell_function(df1)[1]
```

```python
BROKERAGES = {
    "zerodha": {
        "intraday": {
            "buy": {
                "base": 0.0003,
                "stt": 0,
                "exchange txn charges": 0.0000325,
                "gst": 0.00006,
                "sebi": 0.0000005,
                "stamp duty": 0.00003
            },
            "sell": {
                "base": 0.0003,
                "stt": 0.00025,
                "exchange txn charges": 0.0000325,
                "gst": 0.00006,
                "sebi": 0.0000005,
                "stamp duty": 0
            }
        }
    }
}
```

```python
def calculate_brokerage(side, txn_type, price, broker):
    current_brokerages_dict = BROKERAGES[broker][txn_type]
    brokerages = 0
    for head, value in current_brokerages_dict[side].items():
        brokerages += price*value
    return brokerages
```

```python
# get transaction table

# row = -1    # to keep updating row in the transaction table


def get_transctable(transaction_df, time, price, action):
    global row
    row = row + 1    
        
    # setting up intial cols.
    transaction_df.loc[row,'timestamp'] = time


    transaction_df.loc[row,'price'] = price


    transaction_df.loc[row,"buy/sell"] = action


#     transaction_df.reset_index(inplace=True)
#     transaction_df.drop(['index'],axis=1,inplace=True)
    
    transaction_df.loc[row,'qtystock'] = 0


    # for creating brokerage col.
    transaction_df.loc[row,'brokerage'] = calculate_brokerage(transaction_df.loc[row,'buy/sell'],
                                                                'intraday',
                                                                transaction_df.loc[row,'price'],
                                                                'zerodha')

    # for creating qty col.
    if row>0:
        transaction_df.loc[row,'qtystock'] = transaction_df.loc[row-1,'qtystock']
    if transaction_df['buy/sell'][row] == 'buy':
        transaction_df.loc[row,'qtystock'] = transaction_df.loc[row,'qtystock'] + 1
    else:
        transaction_df.loc[row,'qtystock'] = transaction_df.loc[row,'qtystock'] - 1

        # for creating profit col
    if (row%2 != 0):
        if transaction_df.loc[row,'buy/sell'] == 'sell':
            transaction_df.loc[row,'profit'] = transaction_df.loc[row,'price'] - transaction_df.loc[row-1,"price"]
        else:
            transaction_df.loc[row,'profit'] = transaction_df.loc[row-1,'price'] - transaction_df.loc[row,"price"]
    else:
        transaction_df.loc[row,'profit'] = 0


    transaction_df.loc[row,'cum_profit'] = 0.0
    transaction_df.loc[row,'pab'] = 0.0
    transaction_df.loc[row,'cum_pab'] = 0.0

    # for creating cum_profit col.
    if row == 0:
        transaction_df.loc[row,'cum_profit'] = 0
    else:   
        transaction_df.loc[row,'cum_profit'] = transaction_df.loc[row-1,'cum_profit'] + transaction_df.loc[row,'profit']

    # for creating pab col.   
    transaction_df.loc[row,'pab'] = transaction_df.loc[row,'profit'] - transaction_df.loc[row,'brokerage']

    # for creating cum_pab col.
    if row == 0:
        transaction_df.loc[row,'cum_pab'] = transaction_df.loc[row,'pab']
    else:
        transaction_df.loc[row,'cum_pab'] = transaction_df.loc[row-1,'cum_pab'] + transaction_df.loc[row,'pab']
    
    return transaction_df
```
