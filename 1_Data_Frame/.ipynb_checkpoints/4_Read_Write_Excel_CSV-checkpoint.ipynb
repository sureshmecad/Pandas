<h1 style="color:blue" align="left"> 3. Read EXCEL </h1>

```python
import pandas as pd
```

```python
#df = pd.read_excel("C:/Users/deepusuresh/Documents/Data Science/06. Data Sets/Stocks data/Stocks data.csv")
#df

df = pd.read_excel("stocks data.xlsx","Sheet1")
df
```

```python
def convert_people_cell(cell):
    if cell=="n.a.":
        return 'sam walton'
    return cell

def convert_eps_cell(cell):
    if cell=="not available":
        return None
    return cell

df = pd.read_excel("stocks data.xlsx","Sheet1",converters = {
    'people':convert_people_cell,
    'eps':convert_eps_cell
})
df
```

<h1 style="color:blue" align="left"> 4. Writing Excel </h1>

```python
df.to_excel("new.xlsx",sheet_name="stocks",startrow=1,startcol=2)
```

```python
df.to_excel("new.xlsx",sheet_name="stocks",startrow=1,startcol=2,index=False)
```

```python
# Excel file generates in below path (which contains two sheets : stocks / weather)
# C:/Users/deepusuresh/Documents/Data Science/01. Python/Practice/2. PANDAS/stocks_weather.xlsx
```

```python
df_stocks = pd.DataFrame({
    'tickers': ['GOOGL', 'WMT', 'MSFT'],
    'price': [845,65,64],
    'pe': [30.37, 14.26, 30.97],
    'eps': [27.82, 4.61, 2.12]
})

df_weather = pd.DataFrame({
    'day': ['1/1/2017','1/2/2017','1/3/2017'],
    'temperature': [32,35,28],
    'event': ['Rain','Sunny','Snow']
})
```

```python
with pd.ExcelWriter('stocks_weather.xlsx') as writer:
    df_stocks.to_excel(writer,sheet_name="stocks")
    df_weather.to_excel(writer,sheet_name="weather")
```
