### 1. How to lowercase column names in Pandas dataframe
### 2. Apply uppercase to a column in Pandas dataframe

```python
# importing pandas as pd 
import pandas as pd
```

```python
# creating a dataframe

df = pd.DataFrame({'A': ['John', 'bODAY', 'MinA', 'Peter', 'nicky'],'B': ['masters', 'graduate', 'graduate','Masters', 'Graduate'], 
                  'C': [27, 23, 21, 23, 24]}) 
   
df
```

<h3 style="color:blue" align="left"> 1. How to lowercase column names in Pandas dataframe </h3>

### Method : 1

```python
df['A'] = df['A'].str.lower() 
  
df
```

### Method : 2

```python
df['A'].apply(lambda x: x.lower())
```
